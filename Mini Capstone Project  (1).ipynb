Mini Capstone:Predict Student Scores
this Notebook demonstrate how to build a simplelinear regression model to predict the student scores based on hours studied

```python
#import libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, r2_score
```

```python
#loading the data
url="https://raw.githubusercontent.com/AdiPersonalWorks/Random/master/student_scores%20-%20student_scores.csv"
df=pd.read_csv(url)
df.head()
```

```python
# visualizing the data
plt.scatter(df["Hours"], df["Scores"],color='Red')
plt.title("Hours vs Scores")
plt.xlabel("Hours studied")
plt.ylabel("Scores")
plt.grid(True)
plt.show()
```

```python
#preparing data
x=df[["Hours"]]
y=df["Scores"]
```

```python
#Train/Test Split
X_train,X_test, Y_train, Y_test=train_test_split(x,y,test_size=0.2, random_state=42)
```

```python
#creating the model
model=LinearRegression()
model.fit(X_train,Y_train)
```

```python
#predict
Y_pred=model.predict(X_test)
```

```python
#Evaluate model
print("R2 Score:",r2_score(Y_test,Y_pred))
print("Mean Absolute Error:",mean_absolute_error(Y_test,Y_pred))
```

```python
#visualizing the Regression Line
plt.scatter(x,y,color='orange')
plt.plot(x, model.predict(x), color='blue')#Regression Line
plt.xlabel("Hours Studied")
plt.ylabel("Scores")
plt.title("Regression Line-Student Scores")
plt.show()
```

```python
#Predict for custom input
hours=[[19.0]]
predicted_scores=model.predict(hours)
print(f"predicted score for studying {hours[0][0]} hours={predicted_scores[0]:.2f}")
```
